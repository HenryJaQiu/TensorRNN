### visualize the prediction

```python
time_stamp ="05-16-17-13-15-05"
models =["tensor_rnn","basic_lstm","matrix_rnn","tensor_rnn"]
exp='lorenz'

import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt



fig = plt.figure(figsize=(20,3.5))
ax = plt.gca()

seq_id = 1
def plot_ts(ax, model,i):
    data_path = "/var/tmp/tensorRNN/log/"+exp+"/"+time_stamp+"/hz_16/"+model+"/"
    file_name = "predict.npy"
    pred =  np.load(data_path+file_name )
    y_true, y_pred =pred
    ax = fig.add_subplot(1, 4, i+1)
    ax.plot(y_true[:100,seq_id],'g')
    ax.set_ylim([-20,20])
    ax = fig.add_subplot(2, 4, i+1)
    ax.plot(y_pred[:100,seq_id], 'r')
    ax.set_ylim([-20,20])
    ax.set_title(model)
    return pred
i =0
for model in models:
    pred=plot_ts(ax, model,i)
    i += 1
```

```python
y_true, y_pred =pred
plt.plot(y_true[100:200,seq_id])
```
