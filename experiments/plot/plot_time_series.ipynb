### visualize the prediction

```python
time_stamp ="05-16-17-00-05-47"
models =["basic_rnn"]#,"basic_lstm","matrix_rnn","tensor_rnn"]
exp='logistic'

import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt



fig = plt.figure(figsize=(20,3.5))
ax = plt.gca()
seq_id = 1
def plot_ts(ax, model,i):
    data_path = "/var/tmp/tensorRNN/log/"+exp+"/"+time_stamp+"/"+model+"/"
    file_name = "predict.npy"
    pred =  np.load(data_path+file_name )
    y_true, y_pred =pred
    ax = fig.add_subplot(1, 4, i+1)
    ax.plot(y_true[:,seq_id],'g')
    ax.plot(y_pred[:,seq_id], 'r')
    ax.set_title(model)
    return pred
i =0
for model in models:
    pred=plot_ts(ax, model,i)
    i += 1
```

```python
y_true, y_pred =pred
plt.plot(y_true[100:200,seq_id])
```
