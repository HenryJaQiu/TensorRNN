### visualize the prediction

```python
time_stamp ="05-14-17-23-24-30"
models =["basic_rnn","basic_lstm","matrix_rnn","tensor_rnn_einsum"]
exp='U'

import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt



fig = plt.figure(figsize=(20,3.5))
seq_id = 10
def plot_ts(ax, model,i):
    data_path = "/tmp/tensorRNN/log/"+exp+"/"+time_stamp+"/"+model+"/"
    file_name = "predict.npy"
    pred =  np.load(data_path+file_name )
    y_true, y_pred =pred
    ax = fig.add_subplot(1, 4, i+1)
    ax.plot(y_true[:,seq_id],'g')
    ax.plot(y_pred[:,seq_id], 'r')
    ax.set_title(model)
i =0
for model in models:
    plot_ts(ax, model,i)
    i += 1
```
